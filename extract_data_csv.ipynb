```python
import pandas as pd

# Load CSV
df = pd.read_csv("articles.csv", encoding="utf-8")

# --- Label rule (customise here) ---
# Example: mark as 1 if key contains "Included"
df["label"] = df["notes"].apply(
    lambda x: 1 if isinstance(x, str) and "Included" in x else 0
)

# Replace NaN or empty entries with "<Missing>"
df = df.replace(r'^\s*$', pd.NA, regex=True).fillna("<Missing>")

# Save to Excel
df[["title", "abstract", "doi", "label"]].to_excel("parsed_articles.xlsx", index=False)
```
